# Quantum Support Vector Machine Classifier with PennyLane

This notebook implements and compares a quantum variational classifier with a classical SVM using the Iris dataset. The quantum implementation uses PennyLane, a cross-platform library for quantum machine learning.

## 1. Importing Required Libraries

```python
# Basic libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
import sklearn

# Machine learning libraries
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

# PennyLane for quantum computing
import pennylane as qml

# Visualization settings
plt.style.use('default')
sns.set_theme(style="whitegrid")

# Check library versions
print(f"NumPy version: {np.__version__}")
print(f"Pandas version: {pd.__version__}")
print(f"Scikit-learn version: {sklearn.__version__}")
print(f"PennyLane version: {qml.__version__}")
```

## 2. Loading and Preparing the Data

```python
# Loading the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Dataset information
print(f"Dataset shape: {X.shape}")
print(f"Number of classes: {len(np.unique(y))}")
print(f"Classes: {iris.target_names}")
print(f"Features: {iris.feature_names}")

# Normalizing data to [0, 1] range
scaler = MinMaxScaler(feature_range=(0, 1))
features = scaler.fit_transform(X)
labels = y

# Creating a DataFrame for visualization
iris_df = pd.DataFrame(X, columns=iris.feature_names)
iris_df['species'] = pd.Categorical.from_codes(y, iris.target_names)
```

```python
# Visualizing class distribution
plt.figure(figsize=(10, 6))
ax = sns.countplot(data=iris_df, x='species', palette='viridis')
plt.title('Class Distribution in Iris Dataset')
plt.xlabel('Species')
plt.ylabel('Count')

# Adding count labels
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width()/2., p.get_height()),
                ha='center', va='bottom', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Visualizing data with pairplot
plt.figure(figsize=(12, 10))
sns.pairplot(iris_df, hue='species', height=2.5, palette='viridis')
plt.suptitle('Pairplot of Iris Dataset Features', y=1.02, fontsize=16)
plt.tight_layout()
plt.show()
```

## 3. Splitting Data into Training and Testing Sets

```python
# Splitting data with fixed seed for reproducibility
TTS_seed = 123
train_features, test_features, train_labels, test_labels = train_test_split(
    features, labels, random_state=TTS_seed, test_size=0.25, stratify=labels
)

print(f"Training set size: {train_features.shape[0]} samples")
print(f"Testing set size: {test_features.shape[0]} samples")
```

## 4. Classical SVM Classifier Implementation

```python
def train_classic_classifier():
    """Train a classical SVM classifier using One-vs-Rest strategy."""
    svc_classic = OneVsRestClassifier(SVC(kernel="linear", max_iter=100, probability=True))
    svc_classic.fit(train_features, train_labels)
    return svc_classic
```

## 5. Quantum Classifier Implementation with PennyLane

```python
# Number of features (4) defines the number of qubits
num_features = features.shape[1]
num_qubits = num_features

# Create a quantum device
dev = qml.device("default.qubit", wires=num_qubits)

# Define the quantum circuit for feature encoding (similar to ZFeatureMap)
@qml.qnode(dev)
def feature_encoding_circuit(x):
    # Encode features into quantum state
    for i in range(num_qubits):
        qml.RY(np.pi * x[i], wires=i)
    return [qml.expval(qml.PauliZ(i)) for i in range(num_qubits)]

# Visualize the feature encoding circuit with a sample input
sample_input = np.array([0.5, 0.5, 0.5, 0.5])
print("Feature Encoding Circuit:")
print(qml.draw(feature_encoding_circuit)(sample_input))
```

```python
# Define the variational quantum circuit (ansatz)
@qml.qnode(dev)
def variational_circuit(x, weights):
    # Encode features
    for i in range(num_qubits):
        qml.RY(np.pi * x[i], wires=i)
    
    # Variational layers
    for layer in range(2):  # Using 2 repetitions as in the original notebook
        # Rotation gates with trainable parameters
        for i in range(num_qubits):
            qml.RY(weights[layer][i][0], wires=i)
            qml.RZ(weights[layer][i][1], wires=i)
        
        # Entangling gates
        for i in range(num_qubits - 1):
            qml.CNOT(wires=[i, i + 1])
        if num_qubits > 1:  # Connect the last qubit to the first one
            qml.CNOT(wires=[num_qubits - 1, 0])
    
    # Measure the first qubit
    return qml.expval(qml.PauliZ(0))

# Initialize random weights for the variational circuit
np.random.seed(42)
init_weights = np.random.uniform(low=-np.pi, high=np.pi, size=(2, num_qubits, 2))

# Visualize the variational circuit with sample input and weights
print("Variational Circuit (Ansatz):")
print(qml.draw(variational_circuit)(sample_input, init_weights))
```

```python
# Define the cost function for binary classification
def binary_cost(weights, X, y):
    predictions = [variational_circuit(x, weights) for x in X]
    # Convert predictions from [-1,1] to [0,1]
    predictions = [(p + 1) / 2 for p in predictions]
    # Binary cross-entropy loss
    loss = 0
    for pred, target in zip(predictions, y):
        if target == 1:
            loss -= np.log(pred + 1e-10)
        else:
            loss -= np.log(1 - pred + 1e-10)
    return loss / len(X)

# Function to make binary predictions
def binary_predict(weights, X):
    predictions = [variational_circuit(x, weights) for x in X]
    # Convert from [-1,1] to {0,1}
    return [1 if p > 0 else 0 for p in predictions]
```

```python
# Implement One-vs-Rest strategy for multi-class classification
class QuantumClassifier:
    def __init__(self, num_classes):
        self.num_classes = num_classes
        self.weights = [None] * num_classes
        self.callback_values = [[] for _ in range(num_classes)]
    
    def fit(self, X, y):
        # Train a binary classifier for each class
        for class_idx in range(self.num_classes):
            print(f"Training classifier for class {class_idx}...")
            # Convert to binary problem: 1 for current class, 0 for others
            binary_y = [1 if label == class_idx else 0 for label in y]
            
            # Initialize weights
            np.random.seed(42 + class_idx)
            weights = np.random.uniform(low=-np.pi, high=np.pi, size=(2, num_qubits, 2))
            
            # Define optimizer
            opt = qml.GradientDescentOptimizer(stepsize=0.1)
            
            # Optimization loop
            for i in range(100):  # 100 iterations as in the original notebook
                weights, cost = opt.step_and_cost(lambda w: binary_cost(w, X, binary_y), weights)
                self.callback_values[class_idx].append(cost)
                if (i + 1) % 10 == 0:
                    print(f"  Iteration {i+1}: Cost = {cost:.4f}")
            
            self.weights[class_idx] = weights
    
    def predict(self, X):
        # Get scores for each class
        scores = np.zeros((len(X), self.num_classes))
        for class_idx in range(self.num_classes):
            # Get raw predictions from the circuit (values between -1 and 1)
            raw_preds = [variational_circuit(x, self.weights[class_idx]) for x in X]
            # Convert to scores between 0 and 1
            scores[:, class_idx] = [(p + 1) / 2 for p in raw_preds]
        
        # Return class with highest score
        return np.argmax(scores, axis=1)
```

```python
def train_quantum_classifier():
    """Train a quantum classifier using PennyLane."""
    num_classes = len(np.unique(train_labels))
    quantum_classifier = QuantumClassifier(num_classes)
    quantum_classifier.fit(train_features, train_labels)
    return quantum_classifier
```

## 6. Metrics Calculation Function

```python
# Function to calculate specificity
def specificity_score(y_true, y_pred):
    """Calculate average specificity for multi-class classification."""
    cm = confusion_matrix(y_true, y_pred)
    specificities = []
    
    for i in range(len(cm)):
        # True negatives are all elements except those in row i and column i
        true_negatives = np.sum(np.delete(np.delete(cm, i, axis=0), i, axis=1))
        # False positives are elements in column i except element (i,i)
        false_positives = np.sum(cm[:, i]) - cm[i, i]
        
        if true_negatives + false_positives > 0:
            specificities.append(true_negatives / (true_negatives + false_positives))
        else:
            specificities.append(0)
    
    return np.mean(specificities)
```

## 7. Training and Evaluating Models

```python
# Training the classical classifier and measuring time
print("Training the classical classifier...")
start_time = time.time()
classic_classifier = train_classic_classifier()
time_classic = time.time() - start_time

# Making predictions with the classical classifier
train_predicts_classic = classic_classifier.predict(train_features)
test_predicts_classic = classic_classifier.predict(test_features)

print(f"Classical classifier trained in {time_classic:.4f} seconds.")
```

```python
print("Metrics for the classical classifier:")
print("Training time:", time_classic, "seconds")
print("\nClassification report (training):")
print(classification_report(train_labels, train_predicts_classic, digits=4, target_names=iris.target_names))
print("Accuracy (Training):", accuracy_score(train_labels, train_predicts_classic))
print("Specificity (Training):", specificity_score(train_labels, train_predicts_classic))
print("\nClassification report (testing):")
print(classification_report(test_labels, test_predicts_classic, digits=4, target_names=iris.target_names))
print("Accuracy (Testing):", accuracy_score(test_labels, test_predicts_classic))
print("Specificity (Testing):", specificity_score(test_labels, test_predicts_classic))
```

```python
# Training the quantum classifier and measuring time
print("Training the quantum classifier (this may take several minutes)...")
start_time = time.time()
quantum_classifier = train_quantum_classifier()
time_quantum = time.time() - start_time

# Making predictions with the quantum classifier
train_predicts_quantum = quantum_classifier.predict(train_features)
test_predicts_quantum = quantum_classifier.predict(test_features)

print(f"Quantum classifier trained in {time_quantum:.4f} seconds.")
```

```python
print("\nMetrics for the quantum classifier:")
print("Training time:", time_quantum, "seconds")
print("\nClassification report (training):")
print(classification_report(train_labels, train_predicts_quantum, digits=4, target_names=iris.target_names))
print("Accuracy (Training):", accuracy_score(train_labels, train_predicts_quantum))
print("Specificity (Training):", specificity_score(train_labels, train_predicts_quantum))
print("\nClassification report (testing):")
print(classification_report(test_labels, test_predicts_quantum, digits=4, target_names=iris.target_names))
print("Accuracy (Testing):", accuracy_score(test_labels, test_predicts_quantum))
print("Specificity (Testing):", specificity_score(test_labels, test_predicts_quantum))
```

## 8. Confusion Matrix Visualization

```python
# Creating confusion matrix for the classical classifier
cm_classic = confusion_matrix(test_labels, test_predicts_classic)

# Plotting the confusion matrix
plt.figure(figsize=(10, 8))
sns.heatmap(cm_classic, annot=True, fmt="d", cmap="Blues",
            xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.title("Confusion Matrix - Classical Classifier")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()
```

```python
# Creating confusion matrix for the quantum classifier
cm_quantum = confusion_matrix(test_labels, test_predicts_quantum)

# Plotting the confusion matrix
plt.figure(figsize=(10, 8))
sns.heatmap(cm_quantum, annot=True, fmt="d", cmap="Blues",
            xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.title("Confusion Matrix - Quantum Classifier")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()
```

## 9. Performance Comparison

```python
# Comparing performance metrics
metrics = {
    'Model': ['Classical (SVM)', 'Quantum (PennyLane)'],
    'Training Time (s)': [time_classic, time_quantum],
    'Accuracy (Training)': [accuracy_score(train_labels, train_predicts_classic), 
                          accuracy_score(train_labels, train_predicts_quantum)],
    'Accuracy (Testing)': [accuracy_score(test_labels, test_predicts_classic), 
                         accuracy_score(test_labels, test_predicts_quantum)],
    'Specificity (Testing)': [specificity_score(test_labels, test_predicts_classic), 
                              specificity_score(test_labels, test_predicts_quantum)]
}

metrics_df = pd.DataFrame(metrics)
metrics_df
```

## 11. Conclusions

In this notebook, we compared the performance of a classical SVM classifier with a quantum variational classifier implemented using PennyLane on the Iris dataset. Key observations include:

1. **Accuracy**: The classical classifier typically shows slightly higher accuracy than the quantum classifier, both on training and testing data.

2. **Training Time**: The quantum classifier is significantly slower than the classical classifier. This is expected as quantum simulation is computationally intensive.

3. **Classification Patterns**: Both models can perfectly classify the Setosa class (which is linearly separable from the others), but have some difficulty distinguishing between Versicolor and Virginica.

4. **Quantum Potential**: While the quantum classifier may not outperform the classical one on this simple dataset, it demonstrates the potential of quantum computing for machine learning tasks. On more complex problems or larger datasets, quantum algorithms might offer significant advantages.

5. **Current Limitations**: The performance of the quantum classifier is limited by the use of simulators rather than real quantum hardware, as well as the noise and errors inherent in current quantum systems (NISQ era - Noisy Intermediate-Scale Quantum).

6. **Implementation Differences**: This implementation uses PennyLane, which provides a different approach to quantum machine learning compared to Qiskit, with a focus on differentiable quantum computing and hybrid quantum-classical optimization.

This experiment serves as a proof of concept for applying quantum algorithms to classification problems and illustrates both the potential and current challenges of quantum computing in the field of machine learning.
